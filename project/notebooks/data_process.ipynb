```python
import sys
sys.path.append("..")
from src.cleaning import *
from pathlib import Path
import os
from dotenv import load_dotenv
from src.utils import*
from datetime import datetime

load_dotenv()
RAW = Path(os.getenv('DATA_DIR_RAW','../data/raw'))
PROC = Path(os.getenv('DATA_DIR_PROCESSED', '../data/processed'))
```

```python
#load data
df_stock = read_df(f"{RAW}/S&P500_stock_hist_5y.parquet")
df_bench = read_df(f"{RAW}/benchmark_hist_5y.csv")
df_stock
```

```python
df_stock["daily_rtn"] = df_stock.groupby(df_stock["ticker"])["adj close"].pct_change()
df_stock
```

```python
df_bench["daily_rtn"] = df_bench["adj close"].pct_change()
df_bench
```

```python
#drop the missing value here since we do not want to use future data to fill
df_stock = drop_missing(df_stock,columns="daily_rtn")
df_stock.reset_index(drop=True)
```

```python
df_bench = drop_missing(df_bench,columns="daily_rtn")
df_bench.reset_index(drop=True)
```

```python
df_bench["date"]=pd.to_datetime(df_bench["date"])
df_stock['excess_return'] = df_stock['daily_rtn'] - df_stock['date'].map(df_bench.set_index('date')['daily_rtn'])
df_stock
```

```python
df_bench.to_csv(f"{PROC}/bench_hist_5y.csv",index=False)
df_stock.to_parquet(f"{PROC}/S&P500_stock_hist_5y.parquet")
```
